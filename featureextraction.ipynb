```
from skimage import color 
from skimage import io
for img in images:
  grayscale = color.rgb2gray(img) 
  io.imshow(grayscale)
  plt.show()
```

```
x,y=[],[]
for i in images:
    x.append(i[0])
    y.append(i[1])
```

```
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```
fig = plt.figure()
_, axs = plt.subplots(10, 10, figsize=(25,25))
axs = axs.flatten()
for img, ax,k in zip(xtrain, axs,ytrain):
    ax.axis("off")
    ax.set_title(k,fontsize=15)
    ax.imshow(img)
plt.suptitle('Train Data',fontsize=25)
plt.savefig("Training Image.jpg")
plt.show()
```

```
fig = plt.figure()
_, axs = plt.subplots(10, 10, figsize=(25,25))
axs = axs.flatten()
for img, ax,k in zip(xtest, axs,ytest):
    ax.axis("off")
    ax.set_title(k,fontsize=15)
    ax.imshow(img)
plt.suptitle('Testing-Data',fontsize=25)
plt.savefig("Testing Image.jpg")
plt.show()
```
