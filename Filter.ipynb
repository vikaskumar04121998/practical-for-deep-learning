```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Conv2D
import skimage.measure 
from PIL import Image
```

```python
img=Image.open('wolf4.jpg')
plt.imshow(img)
```

```python
rsi = img.resize((np.array((15,15))).astype(int))
image = np.asarray(rsi)
image.shape
```

```python
p=image.reshape(1,15,15,3)
```

```python
def Convolution(p,kernel_size,strides=1, padding='valid'):
 model=Sequential()
 model.add(Conv2D(1, kernel_size=kernel_size,strides=strides,padding=padding,input_shape=(15,15,3)))
 yhat = model.predict(p)
 y=yhat.reshape(yhat.shape[1],yhat.shape[2])
 filtered_image = Image.fromarray(y)
 return filtered_image
```

```python
image_11=Convolution(p,(3,3))
image_12=Convolution(p,(5,5))
image_21=Convolution(p,(3,3),2)
image_22=Convolution(p,(5,5),2)
image_31=Convolution(p,(3,3),1,'same')
image_32=Convolution(p,(5,5),1,'same')
```

```python
plt.imshow(image_11)
plt.title('filter size (3*3)')
entropy = skimage.measure.shannon_entropy(image_11)
print("Entropy:",entropy)
```

```python
plt.imshow(image_12)
plt.title('filter size (5*5)')
entropy = skimage.measure.shannon_entropy(image_12)
print("Entropy:",entropy)
```

```python
plt.imshow(image_21)
plt.title('filter size (3*3), stride=2')
entropy = skimage.measure.shannon_entropy(image_21)
print("Entropy:",entropy)
```

```python
plt.imshow(image_22)
plt.title('filter size (5*5), stride=2')
entropy = skimage.measure.shannon_entropy(image_22)
print("Entropy:",entropy)
```

```python
plt.imshow(image_31)
plt.title('filter size (3*3),stride=1, zero-padding')
entropy = skimage.measure.shannon_entropy(image_31)
print("Entropy:",entropy)
```

```python
plt.imshow(image_32)
plt.title('filter size (5*5),stride=1, zero-padding')
entropy = skimage.measure.shannon_entropy(image_32)
print("Entropy:",entropy)
```
